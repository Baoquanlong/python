## 9.1.2 理解figure与axes


在matplotlib中，绘图有两个基本属于，figure 与 axes，figure 可以想象成绘画的画布，在maplotlib中每一幅画的都是在figure上面绘制的，这个对象包含了所有的子图以及图形元素。  
理解了figure，那么axes 是做什么的呢？假设现在想画一张图，这张图包含了两个散点图水平排列，  
这个时候就要用的`plt.subplots(1,2)`使用子图（也称为axes）来完成操作。利用该函数传建两个对象：figure对象和在figure对象中的axes对象。代码如下：

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(6,4),
sharey=True,dpi=120)

# plt.subplots()函数返回了figure和两个子图的axes对象列表。sahrey = True参数表示两个子图公用一个Y轴，
# dpi参数表示分辨率，子图对象ax1,ax2和plt一样，提供了set_title(),set_xlabel(),set_ylabel()等函数。为了简化代码，可以通过
# 调用axes对象的set()函数一次性完成大部分设置：

ax1.plot([1,2,3,4,5],[1,2,3,6,10],'g*')
ax2.plot([2,5,6,8,9],[5,3,7,9,11],'bD')

ax1.set_title('绿星散点图')
ax2.set_title('蓝钻散点图')
ax1.set_xlabel('X')
ax2.set_ylabel('Y')

ax2.set_xlabel('X')
ax1.set_ylabel('Y')

ax1.set_xlim(0,12)
ax2.set_xlim(0,12)

ax1.set_ylim(0,12)
ax2.set_ylim(0,12)

plt.tight_layout()

ax1.set(title='绿星',xlabel='X',ylabel='Y',xlim=(0,12),ylim=(0,12))
ax1.set(title='蓝钻',xlabel='X',ylabel='Y',xlim=(0,12),ylim=(0,12))
```

## 9.1.3 Matplotlib中面向对象与类Matlab语法的区别

随着对Matplotlib的学习的深入，matplotlib中有两种不同的绘图语法，一种是类似matlab的语法，另一种就是面向对象语法。类matlab语法中所有的绘图使用的plt()都是当前的axes对象。  
所以如果想如前一节一样绘制两幅并列的图，只需要每次先建立一幅字图（使用`plt.plot()`或plt.{其他绘图属性函数}来修改对应的axes就行了。

```python
# 代码如下：
plt.figure(figsize=(6,4),dpi=120)
plt.subplot(1,2,1)  # (行，列，第几个子图)
plt.plot([1,2,3,4,5],[1,2,3,6,10],'g*')
plt.title('绿星散点图')
plt.xlabel('X');plt.ylabel('Y')
plt.xlim(0,12)
plt.ylim(0,12)
# 右边子图
plt.subplot(1,2,2)
plt.plot([2,5,6,8,9],[5,3,7,9,11],'bD')
plt.title('蓝钻散点图')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim(0,12)
plt.ylim(0,12)
plt.show()
```

为了方便设定figure和axes，matplotlib提供了`plt.gca()`和`plt.gcf`函数来获取当前的axes和figures。而`plt.cla()`和`plt.clf()`函数则可清除当前的axes和figure。

```python
## 如果要绘制多幅子图,代码会有大量重复，所以应该采用面向对象的方式。
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import seed,randint

seed(10)
plt.rcParams['font.sans-serif'] = ['SimHei']
# 创建figure,subplots
fig,axes = plt.subplots(2,2,figsize=(10,6),sharex=True,sharey=True,dpi=120)
#定义要正常使用的颜色和标记。
colors = { 0:'g',1:'b',2:'r',3:'y' }
markers = {0:'o',1:'x',2:'*',3:'p'}
# 绘制各子图
for i,ax in enumerate(axes.ravel()):
    ax.plot(sorted(randint(0,10,10)),sorted(randint(0,10,10)),marker=markers[i],color=colors[i])
    ax.set_title('Ax:'+str(i))
    ax.yaxis.set_ticks_position('none')

plt.suptitle('有4副子图的图形： ',verticalalignment='bottom',fontsize=16)
plt.tight_layout()
plt.show()
```

## 9.1.4 修改坐标轴属性
本节先来查看如何修改坐标轴以及标签

* 修改坐标轴的刻度位置以及标签(`plt.xticks()`/`ax.setxticks()`/`ax.setxticklabels()`)  
* 修改坐标轴的刻度显示位置（上/下/左/右)
* 格式刻度标签。  
要设置X轴的刻度和标签，需要用到`xticks()`函数。`plt.xticks()`函数一次性完成了X轴的刻度，标签字体，方向等的设置。`plt.tick_params()`则对每个坐标  
进行了设置，设置了每个轴的(上/下/左/右)都显示朱刻度标记，同事标记显示在内部`(direction='in')`,网格采用蓝色虚线。最后因为要在x轴显示度数，所以需要对X轴的刻度标签进行从弧度到角度的转换。首先利用FuncFormatter定义了一个formatter,之后使用plt.gca(),xaxis,set_major_formatter(formatter)将该formatter应用到X轴。具体代码如下：


```python
from matplotlib.ticker import FuncFormatter
```

```python
plt.rcParams['axes.unicode_minus'] = False
def rad_to_degree(x,pos):
    '''
    角度与弧度的转换
    '''
    return round(x*57.2985,2)
```

```python
plt.figure(figsize=(12,7),dpi=120)
# 在0到2pi之间均匀生成1000个点
X = np.linspace(0,2*np.pi,1000)
plt.plot(X,np.sin(X))
plt.plot(X,np.cos(X))
# 1弧度 = 57.2985度
plt.xticks(ticks=np.arange(0,440/57.2985,90/57.2985),fontsize=12,rotation=30,ha='center',va='top')
# 设置tick参数，每个轴都花刻度，方向是朝内，蓝色网格。
plt.tick_params(axis='both',bottom=True,top=True,left=True,right=True,direction='in',which='major',grid_color='blue')
# x轴的弧度转度
formatter = FuncFormatter(rad_to_degree)
plt.gca().xaxis.set_major_formatter(formatter)
plt.grid(linestyle='--',linewidth=0.5,alpha=0.5)
plt.title('Sin/Cos函数',fontsize=14)
plt.show()
```

单个修改各种属性的确很灵活，但是人工设置各图形属性，一是比较麻烦，另外就是可能使得整个图形存在这样或那样的不一致情况。因此，matplotlib提供了各种设置好的样式供用户选择，通过如下代码，可以查看当前系统中已经安装的样式。

```python
plt.style.available
```

要选择使用那种样式绘图也非常简单，只需要绘图师指定就可以，例如，如下代码就使用了不同样式进行绘图。

```python
import numpy as np
import matplotlib as mpl
mpl.rcParams.update({
    'font.size':18,
    'font.family':'STIXGeneral','mathtext.fontset':'stix'
})

def plot_sine_cosine_wave(style='ggplot'):
    '''
        使用plt.style.use()函数可以选择不同的样式进行绘图。
    '''
    plt.style.use(style)
    plt.figure(figsize=(7,4),dpi=80)
    X = np.linspace(0,2*np.pi,1000)
    plt.plot(X,np.sin(X));plt.plot(X,np.cos(X))
    plt.xticks(ticks = np.arange(0,440/57.2985,90/57.2985),labels=[r'$ 0 $',r'$\frac{\pi}{2}$',r'$\pi$',r'$\frac{3\pi}{2}$'
    ,r'$2\pi$'])
    plt.gca().set(ylim=(-1.25,1.25),xlim=(-0.5,7))
    plt.title(style,fontsize=18)
    plt.show()
```

```python
plot_sine_cosine_wave('seaborn-notebook')
```

## 9.1.6 定制图例，添加标注
绘图过程中经常遇到图例说明，在特定位置添加标注以及文本说明的情况。

图例的设置主要通过`plt.legend()`函数完成

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.style.use(style='seaborn-notebook')
plt.figure(figsize=(10,7),dpi=80)
X = np.linspace(0,2*np.pi,1000)
sine_1 = plt.plot(X,np.sin(X))
cosine_1 = plt.plot(X,np.cos(X))
sine_2 = plt.plot(X,np.sin(X+0.5))
cosine_2 = plt.plot(X,np.cos(X+0.5))
plt.gca().set(ylim = (-1.25,1.5),xlim = (-0.5,7))
plt.title('定制图例',fontsize=18)
plt.legend([sine_1[0],cosine_1[0],sine_2[0],cosine_2[0]],\
    ['sin曲线1','cos曲线1','sin曲线2','cos曲线2'],\
        frameon=False ,      # legend border            
        framealpha=0.2,       #transparency of border
        ncol=2,               # num of columns
        shadow=True,            # shadow on
        borderpad=1,            # thickness of border
        title = 'sin and cos',\
            loc = 'best'
        )
```

使用`plt.text()`和`plt.annotate()`函数可以给推行添加标注以及文本说明，例如如下代码可以完成图形标注功能。

```python
mpl.rcParams.update(mpl.rcParamsDefault)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(14,7),dpi=120)
X = np.linspace(0,8*np.pi,1000)
sine = plt.plot(X,np.sin(X),color='tab:blue')
# 使用Arrayprops和bbox设置标注
t = 57.2985
plt.annotate('峰',xy = (90/57.2985,1.0),xytext = (90/57.2985,1.5),\
    bbox=dict(boxstyle = 'square',fc = 'green',linewidth = 0.1),\
        arrowprops = dict(facecolor = 'green',shrink = 0.01,width = 0.1),\
            fontsize = 12,color = 'white' )
for angle in [440,810,1170]:
    plt.text(angle/t,1.05,str(angle) + "\n度", horizontalalignment='center',color = 'green')
for angle in [270,630,990,1350]:
    plt.text(angle/t,-1.3,str(angle)+"\n度",horizontalalignment='center',color='red')
plt.gca().set(ylim=(-2,2),xlim=(-0.5,26))
plt.title('标注与文本示例',fontsize=18)
plt.show()
```

其中，`plt.annotate()`函数通过xy参数决定标注箭头的位置，`xytext`决定了文本位置，`bbox`指定了文本框的样式，arrowprops制定了箭头的样式，而
`plt.text()`指定了文本的位置。
