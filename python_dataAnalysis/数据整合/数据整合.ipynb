# 8.2 数据合并
# 8.2.1 认识merge操作

在一个数据分析项目中，数据来源于不同格式的文件是一件很常见的事情，pandas提供了merge()函数来完成

```python
import datetime
```

```python
customers = { 'CustomerID':[10,11],
                'Name':['MikeWang','JackMa'],
                'Address':['Address for MikeWang','Address for jackMa']}      
```

```python
customers = pd.DataFrame(customers)
```

```python
customers
```

```python
order= {
    'CustomerID':[10,11,10],
    'OrderDate':[datetime.date(2016,12,1),
    datetime.date(2016,12,1),
    datetime.date(2016,12,2) ]
}
```

```python
order= pd.DataFrame(order)
```

```python
order
```

现在合并两个DataFrame，由于两个表中有相同的CustomerID，所以merge()会自动把CustomerID当作key来合并：

```python
customers.merge(order)
```

上述情况是最简单的合并数据方式，下面看一个更复杂的情形，假设有如下两个数据集：

```python
left_data  = { 'key1':['a','b','c'],
                'key2':['x','y','z'],
                'lvall':[0,1,2] }
                
```

```python
right_data = {
    'key1':['a','b','c'],
    'key2':['x','a','z'],
    'rvall':[6,7,8]
}
```

```python
left= pd.DataFrame(left_data)
```

```python
right = pd.DataFrame(right_data)
```

```python
left
```

```python
right
```

```python
left.merge(right)
```

```python
# 上面的结果只有两行是因为没有指定key，所以自动以两行相同的列为key,如果指定key，则有所不同。
left.merge(right=right,on='key1')
```

这种情况下，因为两列都有key2,所以会自动重命名为key2_x和key2_y

除了按照key进行合并，还可以按照索引进行合并，同样是刚才的数据，得到如下结果。

```python
pd.merge(left,right,left_index=True,right_index=True)
```

## 8.2.2 merge进阶

`merge()`函数还支持通过how参数来指定不同的合并方法。how的参数取值可以有以下四种：
    * inner: 两个DataFrame的key交集，也称内连接;  
    * outer: 两个DataFrame的key并集，也叫外连接;  
    * left : 只是用左边的DataFrame的key，也称左连接；  
    * right: 只使用右边的DataFrame的key，也称右连接；  
使用参数outer，会得到如下结果

```python
left.merge(right,how='outer')
```

```python
# 使用参数left
left.merge(right,how='left')
```

```python
# 使用参数right
left.merge(right,how='right')
```

## 8.2.3 join和concat

```python
## 除了使用merge(),还可以使用join函数来合并
left.join(right,lsuffix='_left',rsuffix='_right')
```

```python
# 下面的代码分别实现了外连接和内连接
left.join(right,how='outer',lsuffix='_l')
```

```python
left.join(right,rsuffix='_r')
```

下面演示concat

```python
df1 = pd.DataFrame(np.arange(9).reshape(3,3),columns=['a','b','c'])
```

```python
df2 = pd.DataFrame(np.arange(9,18).reshape(3,3),columns=['a','b','c'])
```

```python
pd.concat([df1,df2],sort=False)
```

```python
# 如果在concat时使用了key参数，则还可以将key作为level0的索引来
```

```python
pd.concat([df1,df2],keys=['df1','df2'],sort=False)
```
