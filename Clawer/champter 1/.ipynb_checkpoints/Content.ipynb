```python
import requests
from bs4 import BeautifulSoup as BS
```

# 创建一个Content类用于收集网页的一段内容：


```python
class Content:
    def __init__(self,url,title,body) -> None:
        self.url = url
        self.title = title
        self.body = body
```

# 定义解析器函数，对给定的url进行解析，返回content

```python
def getPage(url):
    response = requests.get(url)
    return BS(response.text,'html.parser')
```

```python
def scrapeNYTimes(url):
    bs = getPage(url)
    title = bs.find('h1').text
    lines = bs.find("div",{"id":"mw-content-text"}).find_all("p",recursive=True)
    
    body = " \n ".join([line.text.strip() for line in lines])
    body = body.replace('\xa0',' ')
    body = body.replace('\\n','\n')
    body = body.replace('\\','')
    # body = body.strip()
    return Content(url,title,body)
```

```python
url = r"https://en.wikipedia.org/wiki/Bulgaria"
content = scrapeNYTimes(url)
```
